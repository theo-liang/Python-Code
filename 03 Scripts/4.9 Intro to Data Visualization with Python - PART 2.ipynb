# 01. Importing libraries

```python
# Import libraries

import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
import seaborn as sns
import scipy
```

# 02. Importing Data

```python
# Create path

path = r'C:\Users\16307\Desktop\Tasks - DA Immersion\Instacart Basket Analysis'
```

```python
# Import customers_ords_prods_merge.pkl

customers_ords_prods_merge = pd.read_pickle(os.path.join(path, '02 Data', 'Prepared Data', 'customers_ords_prods_merge.pkl'))
```

# 03. Bar Charts

```python
# Create bar chart for “orders_day_of_week” column

customers_ords_prods_merge['orders_day_of_week'].value_counts().plot.bar()
```

```python
# Sort bar chart by index order rather than largest to smallest

bar = customers_ords_prods_merge['orders_day_of_week'].value_counts().sort_index().plot.bar()
```

```python
# Add color to bar chart

bar = customers_ords_prods_merge['orders_day_of_week'].value_counts().plot.bar(color = ['purple', 'red', 'pink', 'orange', 'yellow', 'green', 'blue'])
```

```python
# Save/Export the bar chart to a file

bar.figure.savefig(os.path.join(path, '04 Analysis','Visualizations', 'bar_orders_dow.png'))
```

# 04. Histograms

```python
# Create histogram for "prices" column

customers_ords_prods_merge['prices'].plot.hist(bins = 25)
```

```python
# Create scatterplot

sns.scatterplot(x = 'prices', y = 'prices',data = customers_ords_prods_merge)
```

# 05. Sampling

```python
# Set a random seed and create a new list "dev" which stores boolean values

np.random.seed(4)
dev = np.random.rand(len(customers_ords_prods_merge)) <= 0.7
```

```python
dev
```

```python
# Produces a list of 10 random numbers between 0 and 1

np.random.rand(10)
```

```python
# Store 70% of the sample in the dataframe big

big = customers_ords_prods_merge[dev]
```

```python
# Store 30% of the sample in the dataframe small

small = customers_ords_prods_merge[~dev]
```

```python
len(customers_ords_prods_merge)
```

```python
len(big) + len(small)
```

```python
# Create a new dataframe, df_2, that contains only the “prices” and “orders_day_of_week” columns from the small subset

df_2 = small[['orders_day_of_week','prices']]
```

# 06. Line Chart

```python
# Create line chart for "orders day of week" and "prices"

line = sns.lineplot(data = df_2, x = 'orders_day_of_week',y = 'prices')
```

# 07. Task

```python
# Create histogram of "order_hour_of_day" column

hist_2 = customers_ords_prods_merge['order_hour_of_day'].plot.hist(bins = 50)
plt.title('Number of Orders by Hour of the Day')
plt.xlabel('Hour of the Day')
plt.ylabel('Number of Orders')
```

The histogram shows the hour of the day on the x-axis and the frequency of orders on the y-axis. From the histogram, we see a relative normal distribution showing that the majority of orders are placed during the middle of the day. 

```python
# Save/Export histogram to a file

hist_2.figure.savefig(os.path.join(path, '04 Analysis','Visualizations', 'hist2_order_hour_of_day.png'))
```
