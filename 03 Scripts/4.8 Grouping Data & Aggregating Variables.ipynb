# #4. Follow the instructions in the Exercise for creating a loyalty flag for existing customers

```python
# Creates a max_order column, dataframe is grouped by user_id, transform function returns maximum on order_number column

ords_prods_merge['max_order'] = ords_prods_merge.groupby(['user_id'])['order_number'].transform(np.max)
```

```python
# Creates new loyalty_flag column and assigns loyalty label corresponding to value

ords_prods_merge.loc[ords_prods_merge['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
ords_prods_merge.loc[(ords_prods_merge['max_order'] <= 40) & (ords_prods_merge['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
ords_prods_merge.loc[ords_prods_merge['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

# #5. Use the loyalty flag you created and check the basic statistics of the product prices for each loyalty category

```python
# Returns descriptive statistic on loyalty_flag

ords_prods_merge.groupby('loyalty_flag').agg({'prices': ['mean']})
```

#5. The prices of products purchased by loyal customers are, on average, lower than the prices of products purchased by new and regular customers.

# #6. Create a spending flag for each user based on the average price across all their orders

```python
# Creates a mean_price_of_products column, dataframe is grouped by user_id, transform function returns mean on prices column

ords_prods_merge['mean_price_of_products'] = ords_prods_merge.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
# Check to see newly added mean_price_of_products column

ords_prods_merge.head(5)
```

```python
# Creates new spending_flag column and assigns spending label corresponding to value

ords_prods_merge.loc[ords_prods_merge['mean_price_of_products'] < 10, 'spending_flag'] = 'Low spender'
ords_prods_merge.loc[(ords_prods_merge['mean_price_of_products'] >= 10), 'spending_flag'] = 'High spender'
```

```python
# Check to see how many low and high spenders

ords_prods_merge['spending_flag'].value_counts(dropna = False)
```

# #7. Create an order frequency flag that marks the regularity of a user’s ordering behavior according to the median in the “days_since_prior_order” column.

```python
# Creates a median_days_between_orders column, dataframe is grouped by user_id, transform function returns median on days_since_prior_order column

ords_prods_merge['median_days_between_orders'] = ords_prods_merge.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
```

```python
# Check to see newly added median_days_between_orders column

ords_prods_merge.head(5)
```

```python
# Creates new order_frequency column and assigns frequency label corresponding to value

ords_prods_merge.loc[ords_prods_merge['median_days_between_orders'] > 20, 'order_frequency'] = 'Non-frequent customer'
ords_prods_merge.loc[(ords_prods_merge['median_days_between_orders'] <= 20) & (ords_prods_merge['median_days_between_orders'] > 10), 'order_frequency'] = 'Regular customer'
ords_prods_merge.loc[ords_prods_merge['median_days_between_orders'] <= 10, 'order_frequency'] = 'Frequent customer'
```

```python
ords_prods_merge['order_frequency'].value_counts(dropna = False)
```
